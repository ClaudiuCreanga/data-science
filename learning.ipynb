```python
colors = ["#02205b", "#1b4690", "#3466ab", "#4f88bc", "#75a8cf", "#a1c5dc", "#c6d8ec",
                "#dee9f5", "#f6fafd", "#fceea5", "#f8d87f", "#f2b05c", "#ed8a4c", "#ed8a4c", "#cf312c", "#a82320", "#6d0f22"]
df['value'] = df['value'].astype(float)
values = df['value'].values
def apply_method(): # 6.9 sec
    cm1 = mcol.ListedColormap(colors)
    norm = matplotlib.colors.Normalize(vmin=np.min(values), vmax=np.max(values), clip=True)
    mapper = cm.ScalarMappable(norm=norm, cmap=cm1)

    return df['value'].apply(lambda row: mcol.to_hex(mapper.to_rgba(row)))
```

```python
df["hex"] = apply_method()
```
