```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

%matplotlib inline
```

```python
def dateparse(date) -> str:
    date = date.split("-")
    return "%s-%s" % (date[0], date[1])
```

```python
data = pd.read_csv(
    "data/kaggle/users.csv",
    parse_dates=["join_date"],
    date_parser=dateparse
)
```
